**Odległość Levenshteina** między dwoma ciągami (np. słowami, zdaniami, tokenami) to minimalna liczba operacji potrzebnych, by przekształcić jeden ciąg w drugi.


```python
def levenshtein_distance(seq1, seq2):
    m, n = len(seq1), len(seq2)
    D = np.zeros((m + 1, n + 1), dtype=int)

    for i in range(m + 1):
        D[i][0] = i
    for j in range(n + 1):
        D[0][j] = j

    for i in range(1, m + 1):
        for j in range(1, n + 1):
            cost = 0 if seq1[i - 1] == seq2[j - 1] else 1
            D[i][j] = min(
                D[i - 1][j] + 1,
                D[i][j - 1] + 1,
                D[i - 1][j - 1] + cost,
            )
    return D[m][n]
```

| **Odwzorowanie logistyczne**                          | **Generacja tekstu (LLM)**                                           | **Uwagi / Metafora chaosu**                                                                     |
| ----------------------------------------------------- | -------------------------------------------------------------------- | ----------------------------------------------------------------------------------------------- |
| (x_n) – obecny stan                                   | (x_n) – bieżąca reprezentacja semantyczna kontekstu                  | Reprezentacja „trajektorii” w przestrzeni semantycznej                                          |

| Iteracja (x*n \to x*{n+1})                            | Iteracja (x*n \to x*{n+1}) przy generowaniu tokenu                   | Każdy nowy token to krok w trajektorii tekstowej                                                |

| Stabilny punkt / atraktor                             | Stały motyw / powtarzalny styl w generowanym tekście                 | Mała wrażliwość na seed, trajektoria pozostaje blisko siebie                                    |
| Chaotyczny reżim                                      | Szybkie „rozbieganie się” tekstów przy minimalnej zmianie seeda      | λ_eff > 0, ekspansja semantyczna, trudna przewidywalność                                        |

### Porównanie Na wykresie Logstycznego i LLMowego


```python
# -------------------------
# Funkcja logistyczna
# -------------------------
def logistic_map(d0, r, steps, d_max):
    d = [d0]
    for k in range(1, steps):
        d_next = r * d[-1] * (1 - d[-1] / d_max)
        d.append(d_next)
    return d


# -------------------------
# Parametry
# -------------------------
d_max = max(len(response_a_split), len(response_b_split))
d0 = trajectory_dist[0] if trajectory_dist[0] > 0 else 1e-6
r_values = [1.5, 2.5, 3.5]  # różne poziomy "chaosu" w odwzorowaniu logistycznym
steps = len(trajectory_dist)

# -------------------------
# Symulacja logistyczna
# -------------------------
logistic_curves = [logistic_map(d0, r, steps, d_max) for r in r_values]

# -------------------------
# Wykres porównawczy
# -------------------------
plt.figure(figsize=(12, 6))

# Rzeczywista trajektoria tekstowa
plt.plot(
    range(1, steps + 1),
    trajectory_dist,
    label="d(k) - rozbieżność tekstów",
    color="black",
    linewidth=2,
)

# Symulacje logistyczne
colors = ["red", "green", "blue"]
for r, curve, color in zip(r_values, logistic_curves, colors):
    plt.plot(
        range(1, steps + 1),
        curve,
        label=f"logistic map r={r}",
        color=color,
        linestyle="--",
    )

plt.xlabel("Długość prefiksu k")
plt.ylabel("Rozbieżność d(k)")
plt.title("Porównanie trajektorii tekstowej z odwzorowaniem logistycznym")
plt.legend()
plt.grid(True)
plt.show()
```
